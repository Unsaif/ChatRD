```python
from utils.synthetic_dataset_utils import get_disease_profile, convert_frequency
from tqdm.auto import tqdm
import numpy as np

# Define the qualitative frequency mapping
qualitative_frequency_map = {
    "Always": 100,
    "Obligate": 100,
    "Very frequent": 95,
    "Frequent": 55,
    "Occasional": 17,
    "Very rare": 3,
    "Unknown": np.nan
}

# Process each disease in filtered_hpo_data
disease_names = []
disease_hpos = []
for disease_name in tqdm(filtered_hpo_data['disease_name'].unique(), total=len(filtered_hpo_data['disease_name'].unique()), desc="Processing Diseases"):
    try:
        disease_profile = get_disease_profile(filtered_hpo_data, disease_name)  # {HPO_ID: Frequency} dictionary
    except:
        print(f"Error processing disease: {disease_name}")
        continue

    # Convert the disease profile dictionary into a DataFrame
    df = pd.DataFrame(disease_profile.items(), columns=['HPO_ID', 'Frequency'])

    # Convert frequency descriptions to numeric values
    df['Frequency'] = df['Frequency'].apply(lambda x: convert_frequency(x, qualitative_frequency_map))
    df['Frequency'] = df['Frequency'].apply(lambda x: qualitative_frequency_map.get(x, x))
    df.fillna({'Frequency': 50}, inplace=True)  # Set default for unknown frequencies if needed

    for _ in range(120):
        sampled_profile = df[df['Frequency'].apply(lambda x: np.random.rand() < (x / 100))]
        disease_names.append(disease_name)
        disease_hpos.append(sampled_profile['HPO_ID'].tolist())
 
# Create the final DataFrame
final_profile_df = pd.DataFrame({'Disease': disease_names, 'HPO_ID': disease_hpos})

# Display the result
final_profile_df
```

```python
import json, numpy as np

# -------- helpers ----------------------------------------------------------
def to_pylist(vec):
    """Return a plain Python list[float] (float32 saves RAM)."""
    if isinstance(vec, np.ndarray):
        return vec.astype(np.float32).tolist()
    return [float(x) for x in vec]

def to_meta(disease, hpo_ids):
    """Metadata values must be scalars â†’ dump the list to JSON text."""
    return {
        "Disease": disease,
        "HPO_IDs": json.dumps(list(hpo_ids))
    }

# -------- streaming ingest -------------------------------------------------
BATCH = 128                       # keep it small â†’ low peak RAM
bucket_ids, bucket_embs = [], []
bucket_docs, bucket_meta = [], []

final_profile_df.dropna(inplace=True)  # drop rows with NaN values

for idx, row in final_profile_df.iterrows():        # row-by-row stream
    bucket_ids.append(str(idx))
    bucket_embs.append(to_pylist(row["Embeddings"]))
    bucket_docs.append(row["Disease"])
    bucket_meta.append(to_meta(row["Disease"], row["HPO_ID"]))

    if len(bucket_ids) == BATCH:                    # flush the batch
        profile_col.add(
            ids=bucket_ids,
            embeddings=bucket_embs,
            documents=bucket_docs,
            metadatas=bucket_meta,
        )
        bucket_ids, bucket_embs, bucket_docs, bucket_meta = [], [], [], []

# flush the tail (if any)
if bucket_ids:
    profile_col.add(
        ids=bucket_ids,
        embeddings=bucket_embs,
        documents=bucket_docs,
        metadatas=bucket_meta,
    )
```

```python
import hnswlib

# Convert embeddings to a NumPy array of type float32
data_embeddings = np.vstack(final_profile_df['Embeddings'].values).astype('float32')

# Set the dimensionality of your embeddings
dim = data_embeddings.shape[1]

# Initialize the index
index = hnswlib.Index(space='cosine', dim=dim)  # Use 'cosine' for cosine similarity

# Estimate the number of elements (for index allocation)
num_elements = data_embeddings.shape[0]

# Initialize the index with desired parameters
index.init_index(max_elements=num_elements, ef_construction=200, M=64)

# Add embeddings to the index
index.add_items(data_embeddings, ids=np.arange(num_elements))

# Set the ef parameter for querying (higher ef leads to better accuracy but slower search)
index.set_ef(200)

# Create a mapping from internal IDs to disease names
id_to_disease = dict(zip(np.arange(num_elements), disease_names))

# Save the index to a file
index.save_index(str(base_path / "hnsw_index_whole_profile.bin"))

# Optionally, save the mapping from IDs to disease names using NumPy
np.save(str(base_path / "id_to_disease_whole_profile.npy"), id_to_disease)
```
